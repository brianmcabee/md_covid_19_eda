# COVID-19 Breakdown by Region

What regions have the worst infection/death/hospitalization rates?

The first analytical question to tackle pertains to the amount of total cases that each region in Maryland has experienced. In order to better examine this question, we need to elaborate on the different regions: Western, Capital, Central, Southern, and Eastern. The Eastern region has the most counties at nine, the Central in second at six counties, followed by Western, Capital, and Southern at three each. In addition, the graph needs elaboration in order to gather a full analysis. The Y axis represents the number of cases while the X axis represents time in months. 

```python
# Plot for cumulative COVID cases by region
df["WESTERN"] = df["GARR"] + df["ALLE"] + df["WASH"]

df["CAPITAL"] = df["FRED"] + df["MONT"] + df["PRIN"]

df["CENTRAL"] = df["ANNE"] + df["BCITY"] + df["BALT"] + df["CARR"] + df["HARF"]+ df["HOWA"]
                
df["SOUTHERN"] = df["CALV"] + df["CHAR"] + df["STMA"]

df["EASTERN"] = df["KENT"] + df["QUEE"] + df["TALB"] + df["CARO"] + df["DORC"]+ df["WICO"] + df["SOME"] + df["WORC"] + df["CECI"]

case_list = [df["WESTERN"], df["CAPITAL"],df["CENTRAL"],df["SOUTHERN"],df["EASTERN"]]

plt.figure(figsize=(15,10))

for x in case_list:
    plt.plot(df["ReportDate"],x,label = x.name)

plt.title('COVID 19 Cases by Region')
plt.xlabel('Report Date')
plt.ylabel('# of Cases')
plt.legend();
```

We can see that the region with the current highest infection rate is Central; although, the Capital region had the most cases for the longest period of time before being overtaken by the Central region. This makes sense as both regions are densely populated urban areas with a lack of space like the rural areas offer. The region with the lowest infection rate is the Western region with a low population and very open, rural areas. 


A following, complimenting graph to the infection rate graph for regions is the regional death rate graph. This visual is also a line graph with the regions being depicted in the same color.

```python
# Plot for cumulative COVID deaths by region

df["WESTERN Deaths"] = df["deathGARR"] + df["deathALLE"] + df["deathWASH"]

df["CAPITAL Deaths"] = df["deathFRED"] + df["deathMONT"] + df["deathPRIN"]

df["CENTRAL Deaths"] = df["deathANNE"] + df["deathBCITY"] + df["deathBALT"] + df["deathCARR"] + df["deathHARF"]+ df["deathHOWA"]
                
df["SOUTHERN Deaths"] = df["deathCALV"] + df["deathCHAR"] + df["deathSTMA"]

df["EASTERN Deaths"] = df["deathKENT"] + df["deathQUEE"] + df["deathTALB"] + df["deathCARO"] + df["deathDORC"]+ df["deathWICO"] + df["deathSOME"] + df["deathWORC"] + df["deathCECI"]

death_list = [df["WESTERN Deaths"], df["CAPITAL Deaths"],df["CENTRAL Deaths"],df["SOUTHERN Deaths"],df["EASTERN Deaths"]]

plt.figure(figsize=(15,10))

for x in death_list:
    plt.plot(df["ReportDate"],x,label = x.name)

plt.title('COVID 19 Deaths by Region')
plt.xlabel('Report Date')
plt.ylabel('# of Cases')
plt.legend();
```

Similarly to the former graph with cases in the regions, the X axis here pertains to time in months; however, the Y axis is now the death rate due to the COVID infection. Interestingly enough, there are some differences. The most significant change is that the Capital area has a higher death rate and has always had a higher death rate than the Central region even though the central region has a higher infection rate. This means that people are more likely to survive the virus on average, if infected, in the central region despite its higher infection rate. 

```python
# Plot for probable COVID deaths by region

df["WESTERN Probable Deaths"] = df["pDeathGARR"] + df["pDeathALLE"] + df["pDeathWASH"]

df["CAPITAL Probable Deaths"] = df["pDeathFRED"] + df["pDeathMONT"] + df["pDeathPRIN"]

df["CENTRAL Probable Deaths"] = df["pDeathANNE"] + df["pDeathBCITY"] + df["pDeathBALT"] + df["pDeathCARR"] + df["pDeathHARF"]+ df["pDeathHOWA"]
                
df["SOUTHERN Probable Deaths"] = df["pDeathCALV"] + df["pDeathCHAR"] + df["pDeathSTMA"]

df["EASTERN Probable Deaths"] = df["pDeathKENT"] + df["pDeathQUEE"] + df["pDeathTALB"] + df["pDeathCARO"] + df["pDeathDORC"]+ df["pDeathWICO"] + df["pDeathSOME"] + df["pDeathWORC"] + df["pDeathCECI"]

pDeath_list = [df["WESTERN Probable Deaths"], df["CAPITAL Probable Deaths"],df["CENTRAL Probable Deaths"],df["SOUTHERN Probable Deaths"],df["EASTERN Probable Deaths"]]

plt.figure(figsize=(15,10))

for x in pDeath_list:
    plt.plot(df["ReportDate"],x,label = x.name)

plt.title('COVID 19 Probable Deaths by Region')
plt.xlabel('Report Date')
plt.ylabel('# of Cases')
plt.legend();
```

# COVID-19 Breakdown By County

What counties have the worst infection/death/hospitalization rates?

While still covering areas of the state of Maryland, the regions can be broken down into their corresponding counties. By examining the following visuals, we can see exactly which counties are causing the average number of COVID cases as well as deaths to rise.

```python
# dataframes for cases and deaths
dfCases = df.loc[:, 'ALLE':'WORC']
dfDeaths = df.loc[:, 'deathALLE':'deathWORC']
dfProbDeaths = df.loc[:, 'pDeathALLE':'pDeathWORC']

# rename columns in dfDeaths dataframe for appropriate county names
for name in dfDeaths:
    dfDeaths = dfDeaths.rename(columns={name : name.replace("death", "")})

# rename columns in dfProbDeaths dataframe for appropriate county names
for nameProb in dfProbDeaths:
    dfProbDeaths = dfProbDeaths.rename(columns={nameProb : nameProb.replace("pDeath", "")})
    
# dataframes for average cases / deaths per county
dfAvgCases = dfCases.mean(axis=0)
dfAvgDeaths = dfDeaths.mean(axis=0)
dfAvgProbDeaths = dfProbDeaths.mean(axis=0)

# dataframes for total cases / deaths per county
dfTotalCases = dfCases.max(axis=0)
dfTotalDeaths = dfDeaths.max(axis=0)
dfTotalProbDeaths = dfProbDeaths.max(axis=0)
```

```python
# show figure for average cases graph
dfAvgCases.plot.bar(rot=0)
plt.title('Maryland COVID-19 Average Cases by County')
plt.xlabel('County')
plt.ylabel('Average COVID-19 Cases')
plt.xticks(rotation=90);
```

```python
# show figure for average deaths graph
dfAvgDeaths.plot.bar(rot=0)
plt.title('Maryland COVID-19 Average Deaths by County')
plt.xlabel('County')
plt.ylabel('Average COVID-19 Deaths')
plt.xticks(rotation=90);
```

```python
# show figure for average probable deaths graph
dfAvgProbDeaths.plot.bar(rot=0)
plt.title('Maryland COVID-19 Average Probable Deaths by County')
plt.xlabel('County')
plt.ylabel('Average COVID-19 Probable Deaths')
plt.xticks(rotation=90);
```

First, let us establish what is seen on the charts. The Y axis contains the average number of cases and deaths, on the latter, while the X axis pertains to the various counties. Now, we can examine the actual figures for the data we need. The county with the highest average cases is Prince George’s county (PRIN) while Montgomery county (MONT) narrowly has more average deaths. This corroborates our previous regional section where the more urban areas are stricken harder by the virus. On the other hand, Garret county, a rural area, has the lowest averages for both cases as well as deaths. What is interesting about these visuals is that Montgomery county tends to have more deaths per case even though they do not have the highest amount of cases. 


Now that we have explored the average cases and deaths per county, we can examine the totals. This seeks to add all of the information since the outbreak to determine which county has been hit the hardest overall rather than which county tends to struggle the most day to day. 

```python
# show figure for total cases graph
dfTotalCases.plot.bar(rot=0)
plt.title('Maryland COVID-19 Total Cases by County')
plt.xlabel('County')
plt.ylabel('Total COVID-19 Cases')
plt.xticks(rotation=90);
```

```python
# show figure for total deaths graph
dfTotalDeaths.plot.bar(rot=0)
plt.title('Maryland COVID-19 Total Deaths by County')
plt.xlabel('County')
plt.ylabel('Total COVID-19 Deaths')
plt.xticks(rotation=90);
```

```python
# show figure for total probable deaths graph
dfTotalProbDeaths.plot.bar(rot=0)
plt.title('Maryland COVID-19 Total Probable Deaths by County')
plt.xlabel('County')
plt.ylabel('Total COVID-19 Probable Deaths')
plt.xticks(rotation=90);
```

Once again, the X axis represents the counties while the Y axis represents the total cases and deaths. Now, these are interesting to look at. It perfectly validates the averages by providing the exact same findings with different, specific values. Once again, we can see that, while Prince George’s county has the most total cases, Montgomery county actually has the most total deaths; additionally, Garret county once again has the lowest for both rates. Therefore, we can conclude, safely, that while Prince George's county has the most cases, Montgomery, another urban area, has more deaths. This is in contrast to the low infection and death rates of rural areas, especially Garret county.

In the examination of infection rates for a given age range, it was important to factor in as many ranges as possible in order to better examine widespread susceptibility. In the presented visual, titled “COVID Cases by age”, we can see age ranges incrementing by nines from zero to 80 and above. The graph also displays a time lapse via the X axis in order to better display how the infection rates have behaved overtime while the Y axis labels the total number of cases. The lines represent the age groups and each is given a unique color to differentiate their approximate infection rate over time. As you can see on the graph, as time has gone on, infection rates across all age ranges have increased to an all time high in November of 2020. As to the age range with the highest cases, the age ranging from 30-39 had the most infection rates for quite some time before being recently overtaken by the 20-29 age range in September of 2020. The lowest infection rate took place in the child age range from 0-9 years old, which consistently held the lowest rate since April. The overall findings seem to indicate that the older the age range, the lower the infection rate; however, two outlying age ranges from 0-9 as well as 10-19, seemed to have very low infection rates as well. That being said, the general tendency of the graph still pointed towards a lower infection rate at higher age ranges.
